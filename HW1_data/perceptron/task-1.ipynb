# Task 1: Perceptron

We train a perceptron (with an offset term) on the 1-vs-5 digit data and
check how well it does on the test set after 1 epoch and after 5 epochs.

```python
import numpy as np

# Load the data. Each row is: symmetry, intensity, label.
train = np.loadtxt('train_1_5.csv', delimiter=',')
test = np.loadtxt('test_1_5.csv', delimiter=',')

X_train = train[:, :2]
y_train = train[:, 2]
X_test = test[:, :2]
y_test = test[:, 2]


def perceptron(X, y, epochs):
    # Perceptron with an offset. We go through the points in order
    # (no shuffling) and only update when a point is misclassified.
    theta = np.zeros(X.shape[1])
    theta_0 = 0.0
    for e in range(epochs):
        for i in range(len(X)):
            xi = X[i]
            yi = y[i]
            if yi * (np.dot(theta, xi) + theta_0) <= 0:
                theta = theta + yi * xi
                theta_0 = theta_0 + yi
    return theta, theta_0


def accuracy(X, y, theta, theta_0):
    # fraction of points we classify correctly
    preds = np.sign(np.dot(X, theta) + theta_0)
    return np.mean(preds == y)
```

```python
# (a) one pass through the data
theta1, theta0_1 = perceptron(X_train, y_train, epochs=1)
acc1 = accuracy(X_test, y_test, theta1, theta0_1)

print('After 1 epoch:')
print('theta  =', theta1)
print('offset =', theta0_1)
print('test accuracy =', round(acc1, 4))
```

```python
# (b) five passes through the data
theta5, theta0_5 = perceptron(X_train, y_train, epochs=5)
acc5 = accuracy(X_test, y_test, theta5, theta0_5)

print('After 5 epochs:')
print('theta  =', theta5)
print('offset =', theta0_5)
print('test accuracy =', round(acc5, 4))
```
